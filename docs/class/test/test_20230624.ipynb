## 제 1유형

### 문제 1) 
주어진 데이터를 이용하여 이동시간(출동시간과 도착시간의 차이)이 평균적으로 가장 오래 걸린 구급대의 이동시간을 구하시오. (분 단위로 환산 및 반올림하여 정수로 출력)

```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_1.csv')
df.head()
```

```python
df.info()
```

```python
df["출동시간"] = pd.to_datetime(df["출동시간"])
df["도착시간"] = pd.to_datetime(df["도착시간"])
```

```python
df["차이"] = df["도착시간"] - df["출동시간"]
df['차이'] = df['차이'].dt.total_seconds()/60
df.head()
```

```python
grouped = df.groupby('구급대')['차이'].mean()
grouped
```

```python
result = grouped.max()
print(round(result))
```

### 문제2)
주어진 데이터를 이용하여 교수 1인당 맡고 있는 학생 수가 가장 많은 대학교를 찾고, 그 대학교의 전체 교수의 수를 구하시오.(정수로 출력)

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_2.csv')
df.head()
```

```python
df['학생수'] = df['경영학과'] + df['경제학과'] + df['컴퓨터공학과'] + df['산업공학과'] + df['국문학과'] + df['영문학과']
df_sort = df[['대학교명', '교수', '학생수']]
df_sort.head()
```

```python
df_sort['교수당 학생수'] = round(df['학생수'] / df['교수'])
df_sort.head()
```

```python
maxidx = df_sort['교수당 학생수'].idxmax()
print(maxidx)
```

```python
result = df.loc[maxidx, '교수']
print(result)
```

### 문제3)
주어진 데이터를 이용하여 연도별 총 범죄 건수의 월평균 값을 구하고, 그 값이 가장 큰 연도의 총 범죄 건수의 월평균 값을 집계하시오.(반올림하여 정수로 출력)

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_3.csv')
df.head()
```

```python
df.info()
```

```python
df['total'] = df.iloc[:, 1:-1].sum(axis=1)
df
```

```python
df.head()
```

```python
drop_column = df.iloc[:, 1:-1].columns
df.drop(drop_column, axis=1, inplace=True)
```

```python
df.head()
```

```python
df['연도'] = df['날짜'].str[:4]
df.head()
```

```python
grouped = df.groupby('연도')['total'].mean()
```

```python
max(round(grouped))
```
