# 차원 축소(Dimensionality Reduction)
- 데이터를 간단하게 줄이되, 중요한 정보는 최대한 보존하는 기술

## 차원(Dimension) : 데이터의 특징(변수, 피처, 컬럼)
- 차원의 저주 : 차원이 많을 때 생기는 문제점
    - 시각화 어려움 : 3차원 이상은 눈으로 보기 어려움
    - 연산 속도 느림 : 고차원은 계산 비용 증가
    - 모델 성능 저하 : 중요하지 않은 피처가 노이즈로 작용

### Feature Selection(특징 선택)
- 중요한 피처만 골라서 사용
- SelectKBest, RFE 등

### Feature Extraction(특징 추출)
- 기존 피처들을 조합해서 새로운 저차원 특징 생성
- PCA, LDA, t-SNE 등

```python
# PCA 주성분 분석
## 데이터를 잘 설명할 수 있는 새로운 축(=주성분)을 만들어서,
## 원래의 고차원 데이터를 이 새로운 축 기준으로 줄이는 것

from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 준비
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# PCA 변환 (2차원으로 축소)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 시각화
plt.figure(figsize=(12, 5))

# 원본 데이터 (sepal length, sepal width만 사용)
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.xlabel(feature_names[0])  # sepal length
plt.ylabel(feature_names[1])  # sepal width
plt.title('📌 원본 데이터 (2 features)')
plt.grid(True)

# PCA 변환된 데이터
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('📌 PCA 차원 축소 결과')
plt.grid(True)

plt.tight_layout()
plt.show()
```
